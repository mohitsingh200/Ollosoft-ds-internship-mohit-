```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
reliance = yf.download(
    "RELIANCE.NS",
    period="2y"
)

reliance.head()
```

```python
print(reliance.shape)
print(reliance.columns)
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    reliance.index,
    reliance["Close"],
    label="Reliance Close Price"
)
```

```python
plt.title("Reliance Closing Price (Last 2 Years)")
```

```python
plt.xlabel("Date")
plt.ylabel("Closing Price")
```

```python
plt.legend()
```

```python
plt.grid(True)
```

```python
plt.show()
```
