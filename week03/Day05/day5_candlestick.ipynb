# Candlestick Charts

## Objective

Create a candlestick chart for Reliance stock using the last 3 months of data.

### Features
- Candlestick chart
- 20-Day Moving Average
- 50-Day Moving Average
- Interactive Plotly chart

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

## Download Last 3 Months Data

Download Reliance stock data using yfinance.

```python
reliance = yf.download(
    "RELIANCE.NS",
    period="3mo",
    auto_adjust=False
)

reliance.head()
```

```python
print(reliance.columns)
```

```python
reliance.columns = reliance.columns.get_level_values(0)
```

## Calculate Moving Averages

Moving averages help identify market trends.

```python
reliance["MA20"] = (
    reliance["Close"]
    .rolling(20)
    .mean()
)

reliance["MA50"] = (
    reliance["Close"]
    .rolling(50)
    .mean()
)

reliance.head()
```

```python
fig = go.Figure()
```

```python
fig.add_trace(
    go.Candlestick(
        x=reliance.index,
        open=reliance["Open"],
        high=reliance["High"],
        low=reliance["Low"],
        close=reliance["Close"],
        name="Reliance"
    )
)
```

```python
fig.add_trace(
    go.Scatter(
        x=reliance.index,
        y=reliance["MA20"],
        mode="lines",
        name="20-Day MA"
    )
)
```

```python
fig.add_trace(
    go.Scatter(
        x=reliance.index,
        y=reliance["MA50"],
        mode="lines",
        name="50-Day MA"
    )
)
```

```python
fig.update_layout(
    title="Reliance Candlestick Chart",
    xaxis_title="Date",
    yaxis_title="Price",
    template="plotly_white",
    xaxis_rangeslider_visible=False
)
```

```python
fig.show()
```

```python
fig.write_html(
    "../images/reliance_candlestick.html"
)
```

## Observations

1. Candlestick charts show daily price movement.
2. Green candles indicate price increase.
3. Red candles indicate price decrease.
4. Moving averages smooth short-term fluctuations.
5. MA20 reacts faster than MA50.
