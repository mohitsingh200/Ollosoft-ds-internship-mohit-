# Interactive Charts with Plotly

## Objective

Create an interactive stock price chart using Plotly.

### Features
- Interactive zoom and pan
- Hover tooltip
- Daily change calculation
- Professional financial visualization

```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
```

## Download Reliance Stock Data

Download 2 years of historical stock data using yfinance.

```python
reliance = yf.download(
    "RELIANCE.NS",
    period="2y",
    auto_adjust=True
)

reliance.head()
```

```python
print(reliance.columns)
```

```python
reliance.columns = reliance.columns.get_level_values(0)
```

```python
print(reliance.columns)
```

## Calculate Daily Percentage Change

Daily percentage change helps measure day-to-day stock movement.

```python
reliance["Daily_Change_%"] = (
    reliance["Close"]
    .pct_change()
    * 100
)

reliance.head()
```

```python
fig = px.line(
    reliance,
    x=reliance.index,
    y="Close",
    title="Reliance Closing Price (Interactive)"
)
```

```python
fig.update_traces(
    customdata=reliance[["Daily_Change_%"]],
    hovertemplate=
    "<b>Date:</b> %{x}<br>" +
    "<b>Price:</b> ₹%{y:.2f}<br>" +
    "<b>Daily Change:</b> %{customdata[0]:.2f}%<br>" +
    "<extra></extra>"
)
```

```python
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Closing Price",
    hovermode="x unified",
    template="plotly_white"
)
```

```python
fig.show()
```

```python
fig.write_html(
    "../images/reliance_plotly_chart.html"
)
```

## Observations

1. Plotly provides interactive visualizations.
2. Hover tooltips display detailed stock information.
3. Daily percentage change helps identify market volatility.
4. Interactive charts improve data exploration.
