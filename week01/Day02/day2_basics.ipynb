# SIMPLE INTEREST

```python
p = 1000
r = 5
t = 2

si = (p * r * t) / 100

print("Simple Interest =", si)
```

# COMPOUND INTEREST

```python
p = 1000
r = 5
t = 2

amount = p * ((1 + r / 100) ** t)
ci = amount - p

print("Compound Interest =", ci)
```

# simple interest and compound interest for 3 different inputs.

```python
values = [
    (1000, 5, 2),
    (5000, 7, 3),
    (10000, 10, 5)
]

for p, r, t in values:
    si = (p * r * t) / 100
    amount = p * ((1 + r / 100) ** t)
    ci = amount - p

    print("Principal:", p)
    print("Simple Interest:", si)
    print("Compound Interest:", ci)
    
```
