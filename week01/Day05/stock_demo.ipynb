```python
from stock import Stock
```

```python
stock1 = Stock(
    "RELIANCE",
    [2400, 2450, 2500, 2480, 2550]
)
```

```python
stock2 = Stock(
    "TCS",
    [3200, 3180, 3250, 3300, 3280]
)
```

```python
print("Ticker:", stock1.ticker)

print("Average Price:", stock1.avg_price())

print("Total Return:", stock1.total_return())

print("Max Drawdown:", stock1.max_drawdown())
```

```python
print("Ticker:", stock2.ticker)

print("Average Price:", stock2.avg_price())

print("Total Return:", stock2.total_return())

print("Max Drawdown:", stock2.max_drawdown())
```
