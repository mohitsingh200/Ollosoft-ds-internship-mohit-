```python
prices = np.array([100, 105, 110, 108, 120])
```

```python
returns = (prices[1:] - prices[:-1]) / prices[:-1]

print(returns)
```

```python
returns_percent = returns * 100

print(returns_percent)
```

```python
prices = [100,105,110,108,120]

returns = []

for i in range(1, len(prices)):
    
    r = (prices[i] - prices[i-1]) / prices[i-1]
    
    returns.append(r)

print(returns)
```
