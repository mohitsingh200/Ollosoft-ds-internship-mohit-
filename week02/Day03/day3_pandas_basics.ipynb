```python
import pandas as pd
import yfinance as yf
```

```python
reliance = yf.download("RELIANCE.NS", period="2y")
reliance["Stock"] = "RELIANCE"

tcs = yf.download("TCS.NS", period="2y")
tcs["Stock"] = "TCS"

infy = yf.download("INFY.NS", period="2y")
infy["Stock"] = "INFY"
```

```python
reliance.reset_index(inplace=True)
tcs.reset_index(inplace=True)
infy.reset_index(inplace=True)
```

```python
df = pd.concat([reliance, tcs, infy])

df.head()
```
