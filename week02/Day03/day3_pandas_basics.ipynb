```python
import pandas as pd
import yfinance as yf
```

```python
reliance = yf.download(
    "RELIANCE.NS",
    period="2y"
)

reliance.head()
```

```python
tcs = yf.download(
    "TCS.NS",
    period="2y"
)

tcs.head()
```

```python
infy = yf.download(
    "INFY.NS",
    period="2y"
)

infy.head()
```

```python
reliance["Stock"] = "RELIANCE"

tcs["Stock"] = "TCS"

infy["Stock"] = "INFY"
```

```python
stocks = pd.concat([
    reliance,
    tcs,
    infy
])
```
