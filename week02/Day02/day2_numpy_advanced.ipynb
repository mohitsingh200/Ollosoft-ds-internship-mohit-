```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
daily_returns = np.random.normal(
    loc=0.001,
    scale=0.02,
    size=1000
)

print(daily_returns[:10])
```

```python
print(np.mean(daily_returns))
```

```python
print(np.std(daily_returns))
```

```python
print("25th:", np.percentile(daily_returns,25))

print("50th:", np.percentile(daily_returns,50))

print("75th:", np.percentile(daily_returns,75))
```

```python
start_price = 100
```

```python
price_series = start_price * np.cumprod(
    1 + daily_returns
)

print(price_series[:10])
```

```python
print("Average Price:",
      np.mean(price_series))

print("Highest Price:",
      np.max(price_series))

print("Lowest Price:",
      np.min(price_series))

print("Final Price:",
      price_series[-1])
```
