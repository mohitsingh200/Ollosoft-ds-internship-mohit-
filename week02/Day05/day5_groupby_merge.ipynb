```python
import pandas as pd
```

```python
df = pd.read_csv("../Day04/stocks_raw.csv")

df.head()
```

```python
print(df.columns)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df.info()
```

```python
sector_map = {
    "RELIANCE": "Energy",
    "TCS": "Information Technology",
    "INFY": "Information Technology"
}

df["Sector"] = df["Stock"].map(sector_map)

df.head()
```

```python
df.info()
```

```python
df["Close"] = pd.to_numeric(df["Close"], errors="coerce")
```

```python
print(df["Close"].dtype)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
monthly_avg = (
    df.groupby([
        "Stock",
        pd.Grouper(key="Date", freq="ME")
    ])["Close"]
    .mean()
    .reset_index()
)

monthly_avg.head()
```

```python
monthly_volatility = (
    df.groupby([
        "Stock",
        pd.Grouper(key="Date", freq="ME")
    ])["Close"]
    .std()
    .reset_index()
)

monthly_volatility.columns = [
    "Stock",
    "Date",
    "Monthly_Volatility"
]

monthly_volatility.head()
```

```python
quarterly = (
    df.groupby([
        "Stock",
        pd.Grouper(key="Date", freq="QE")
    ])["Close"]
    .mean()
    .reset_index()
)

quarterly.head()
```

```python
monthly_summary = pd.merge(
    monthly_avg,
    monthly_volatility,
    on=["Stock", "Date"]
)

monthly_summary.to_csv(
    "monthly_summary.csv",
    index=False
)

print("File Saved Successfully")
```
