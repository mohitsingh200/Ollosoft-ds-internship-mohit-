```python
import pandas as pd
```

```python
df = pd.read_csv("stocks_raw.csv")
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.info()
```

```python
df["Close"] = pd.to_numeric(df["Close"], errors="coerce")
```

```python
print(df["Close"].max())
```

```python
reliance = df[df["Stock"] == "RELIANCE"]

reliance.head()
```

```python
reliance_above_1500 = reliance[reliance["Close"] > 1500]

reliance_above_1500.head()
```

```python
reliance["Pct_Change"] = reliance["Close"].pct_change() * 100

reliance.head()
```

```python
reliance["Abs_Change"] = reliance["Pct_Change"].abs()

reliance.head()
```

```python
sorted_days = reliance.sort_values(
    by="Abs_Change",
    ascending=False
)

sorted_days.head(10)
```

```python
top10 = reliance.nlargest(
    10,
    "Abs_Change"
)

top10
```

```python
tcs = df[df["Stock"] == "TCS"].copy()

tcs["Pct_Change"] = tcs["Close"].pct_change(fill_method=None) * 100

tcs["Abs_Change"] = tcs["Pct_Change"].abs()

tcs.nlargest(10, "Abs_Change")
```

```python
infy = df[df["Stock"] == "INFY"].copy()

infy["Pct_Change"] = infy["Close"].pct_change(fill_method=None) * 100

infy["Abs_Change"] = infy["Pct_Change"].abs()

infy.nlargest(10, "Abs_Change")
```

Daily Percentage Change
