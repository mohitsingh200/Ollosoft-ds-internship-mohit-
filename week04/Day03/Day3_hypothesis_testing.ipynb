# Hypothesis Testing on Stock Returns

## Objectives

1. Test whether Reliance mean daily return differs from zero.
2. Test whether Reliance and TCS have different mean returns.
3. Interpret p-values and make decisions.

```python
import pandas as pd
import numpy as np

import yfinance as yf

from scipy.stats import ttest_1samp
from scipy.stats import ttest_ind
```

```python
reliance = yf.download(
    "RELIANCE.NS",
    period="5y",
    auto_adjust=True
)

tcs = yf.download(
    "TCS.NS",
    period="5y",
    auto_adjust=True
)
```

```python
if isinstance(reliance.columns, pd.MultiIndex):
    reliance.columns = reliance.columns.get_level_values(0)

if isinstance(tcs.columns, pd.MultiIndex):
    tcs.columns = tcs.columns.get_level_values(0)
```

```python
reliance_returns = (
    reliance["Close"]
    .pct_change()
    .dropna()
)

tcs_returns = (
    tcs["Close"]
    .pct_change()
    .dropna()
)
```

## One-Sample t-Test

### H₀:
Mean daily return of Reliance = 0

### H₁:
Mean daily return of Reliance ≠ 0

```python
t_stat, p_value = ttest_1samp(
    reliance_returns,
    popmean=0
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

## Interpretation

If p-value < 0.05:

We reject the null hypothesis and conclude that the mean daily return of Reliance is significantly different from zero.

Otherwise:

There is insufficient evidence to conclude that the mean daily return differs from zero.

## Two-Sample t-Test

### H₀:
Mean return of Reliance = Mean return of TCS

### H₁:
Mean return of Reliance ≠ Mean return of TCS

```python
t_stat2, p_value2 = ttest_ind(
    reliance_returns,
    tcs_returns,
    equal_var=False
)

print("T-statistic:", t_stat2)
print("P-value:", p_value2)
```

```python
alpha = 0.05

if p_value2 < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
