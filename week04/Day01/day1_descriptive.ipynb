# Descriptive Statistics of Stock Returns

## Objective

Analyze daily returns of five Indian stocks using descriptive statistics.

Statistics computed:

- Mean
- Median
- Mode
- Variance
- Standard Deviation
- Skewness
- Kurtosis

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
stocks = [
    "RELIANCE.NS",
    "TCS.NS",
    "INFY.NS",
    "HDFCBANK.NS",
    "ICICIBANK.NS"
]
```

```python
data = yf.download(
    stocks,
    period="5y",
    auto_adjust=True
)
```

```python
close_prices = data["Close"]

close_prices.head()
```

```python
returns = close_prices.pct_change().dropna()

returns.head()
```

## Daily Returns

Daily returns are calculated using percentage change in closing prices.

```python
summary = []
```

```python
for stock in returns.columns:

    r = returns[stock]

    summary.append({
        "Stock": stock,
        "Mean": r.mean(),
        "Median": r.median(),
        "Mode": r.mode().iloc[0],
        "Variance": r.var(),
        "Std_Dev": r.std(),
        "Skewness": r.skew(),
        "Kurtosis": r.kurt()
    })
```

```python
summary_df = pd.DataFrame(summary)

summary_df
```

```python
summary_df = summary_df.round(4)

summary_df
```

```python
summary_df.to_csv(
    "descriptive_statistics.csv",
    index=False
)
```

# Interpretation of Skewness

## Skewness measures the asymmetry of a distribution.

 Positive Skewness:
  The distribution has a longer right tail. Large positive returns occur occasionally.

 Negative Skewness:
  The distribution has a longer left tail. Large negative returns occur occasionally.

 Near Zero:
  Returns are approximately symmetric.

# Interpretation of Kurtosis

## Kurtosis measures the heaviness of tails.

 High Kurtosis:
  More extreme returns than a normal distribution. Indicates higher risk.

 Low Kurtosis:
  Fewer extreme returns.

 Kurtosis near Zero:
  Similar to a normal distribution.
