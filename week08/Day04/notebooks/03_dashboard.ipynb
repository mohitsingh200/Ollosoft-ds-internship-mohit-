```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
from plotly.subplots import make_subplots

from ipywidgets import interact, widgets

import joblib
```

```python
data = pd.read_csv("../data/features.csv")

data.head()
```

```python
print(data.columns.tolist())
```

```python
model = joblib.load("../models/best_model.joblib")
```

```python
feature_columns = [
    "SMA_10",
    "SMA_20",
    "SMA_50",
    "EMA_10",
    "EMA_20",
    "RSI",
    "MACD",
    "Lag_1",
    "Lag_5",
    "Lag_21"
]
```

```python
stocks = ["RELIANCE.NS"]
```

```python
def dashboard(stock):

    # df = data[data["Ticker"] == stock].copy()
    df = data.copy()

    latest = df.iloc[-1]

    X_latest = latest[feature_columns].values.reshape(1, -1)

    prediction = model.predict(X_latest)[0]

    fig = make_subplots(
        rows=3,
        cols=1,
        shared_xaxes=True,
        vertical_spacing=0.05,
        subplot_titles=(
            "Closing Price",
            "RSI",
            "MACD"
        )
    )

    # Close
    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["Close"],
            name="Close"
        ),
        row=1,
        col=1
    )

    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["SMA_20"],
            name="SMA 20"
        ),
        row=1,
        col=1
    )

    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["EMA_20"],
            name="EMA 20"
        ),
        row=1,
        col=1
    )

    # RSI

    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["RSI"],
            name="RSI"
        ),
        row=2,
        col=1
    )

    # MACD

    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["MACD"],
            name="MACD"
        ),
        row=3,
        col=1
    )

    fig.add_trace(
        go.Scatter(
            x=df["Date"],
            y=df["Signal"],
            name="Signal"
        ),
        row=3,
        col=1
    )

    fig.update_layout(
        height=900,
        width=1200,
        title=f"{stock} Dashboard"
    )

    fig.show()

    print("="*50)

    print("Stock :", stock)

    print()

    print("Model Prediction")

    if prediction == 1:
        print("📈 Next Day : UP")
    else:
        print("📉 Next Day : DOWN")

    print()

    print("Historical Accuracy")

    if "Prediction" in df.columns:

        acc = (df["Prediction"] == df["Target"]).mean()*100

        print(f"{acc:.2f}%")

    else:

        print("Prediction column not available.")

    print("="*50)
```

```python
interact(
    dashboard,
    stock=widgets.Dropdown(
        options=stocks,
        description="Stock:"
    )
);
```
