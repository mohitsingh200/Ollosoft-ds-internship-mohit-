```python
# Data manipulation
import pandas as pd
import numpy as np

# Stock data
import yfinance as yf
```

## Download 5 Years of Stock Data

```python
# Indian Large Cap Stocks

tickers = [
    "RELIANCE.NS",
    "TCS.NS",
    "INFY.NS",
    "HDFCBANK.NS",
    "ICICIBANK.NS"
]
```

```python
# Download historical data

data = yf.download(

    tickers,

    start="2019-01-01",

    end="2024-12-31",

    auto_adjust=True,

    group_by="ticker"
)

# Display first rows

data.head()
```

## Save Raw Dataset

```python
# Save raw stock data

data.to_csv("../data/raw_stock_data.csv")

print("Dataset saved successfully.")
```
